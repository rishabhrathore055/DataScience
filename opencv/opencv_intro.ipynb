## Working with Opencv & Images


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('ninja_zx_10r.jpg')
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
plt.imshow(rgb_img)
plt.axis("off")
plt.show()
```

```python
crop_image = rgb_img[50:800,200:1100:]
plt.imshow(crop_image)
plt.axis("off")
plt.show()
```

```python
cv2.imwrite('savedImage.jpg',crop_image)
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0] * crop_image.shape[1]
total_pixels
```

```python
# Resize
test = cv2.resize(crop_image,(100,100))
```

```python
plt.imshow(test)
```

```python
cv2.imwrite('testImage.jpg',test)
```
