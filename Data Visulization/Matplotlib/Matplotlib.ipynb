```python
# adjust the size of any plot
plt.figure(figsize=(10,5))
plt.scatter(x,y1,color = 'b',label = "apple",marker ='^')
plt.scatter(x,y2,color = 'black',label = "Kiwi")
plt.xlabel("time")
plt.ylabel("price")
plt.title("Time vs price")
plt.legend()
plt.show()
```

# Bar graphs

```python
x_coordinates = np.array([2,4,5])*2
plt.bar(x_coordinates - 0.5,[15,18,16],width = 0.5,label = "Current year",tick_label =["Gold","Platinum","Silver"],color = "blue")
plt.bar(x_coordinates + 0.5,[20,14,10],width = 0.5,label = "next year",color = "orange")
plt.title("Metal price comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,25)
plt.xlim(3,12)
plt.legend()
plt.show()
```

# Histogram

```python
xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(xsn * sigma + u)
X2 = np.round(xsn * 7 + 40)
print(X)
```

```python
plt.hist(X,alpha=.8,label="Maths")
plt.hist(X2,alpha=.8,label="Physics")
plt.xlabel("Marks Range")
plt.ylabel("FreqCount of students")
plt.title("His")
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3],[2,4,9])
plt.show()
```

```python
x = np.arange(10)
y = x ** 2

x2 = np.arange(-5,15)
y2 = x2 + 50

plt.title("Squares")
plt.xlabel("X-values")
plt.ylabel("squares values")

plt.xlim(-1,15)
plt.ylim(-20,120)

plt.plot(x,y,'r--',label = 'square line')
plt.plot(x2,y2,'r--',label = 'linear line')

plt.legend()

plt.show()
```

```python
plt.plot(x,y,'b--',x2,y2,'ro')
plt.show()
```

```python
import math

x = np.arange(1,101)

# y = np.log(x)
# y = [math.log(i) for i in x ]
y = list(map(lambda i : math.log(i),x))
y = list(map(math.log,x))
y = np.vectorize(math.log)(x)

plt.plot(x,y)
plt.show()
```

```python
import math 

sigmod = lambda x : 1 /(1 + np.exp(-x))

x = np.arange(-10,10)

y = [sigmod(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,10,0.1)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x)

plt.figure(figsize=(15,10))

# num row, num col, subplot num
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)


plt.show()
```
